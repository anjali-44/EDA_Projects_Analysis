2. popularity based system

while building we will try to incorporate 1 and 2 

1. data
2. data pre processing 
3. feature extraction
4. user input (their fav movie)
   based on this we will suggest them which movie to watch
5. for that we use cosine similarity algo (to find      similarity between vectors)
6. suggest list of movies

here we are not using any ML model we will use cosine similarity algorithm (very much useful in finding the relationship between differnt data points) as it will give better results in this case


```python
import pandas as pd
import numpy as np
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# data collection and processing
```

```python
movies_data = pd.read_csv('movies.csv')
```

```python
movies_data.head()
```

```python
movies_data.shape
```

```python
# selecting the relevant features for recommendation 
# for building content and popularity based movies

selected_features = ['genres', 'keywords', 'tagline', 'cast', 'director']
selected_features
```

```python
# replacing the null values with null string

for feature in selected_features:
    movies_data[feature] = movies_data[feature].fillna('')
```

```python
# combining all the 5 selected feature

combined_feature = movies_data['genres']+' '+movies_data['keywords']+' '+movies_data['tagline']+' '+movies_data['cast']+' '+movies_data['director']
```

```python
combined_feature
```

```python
# converting the text data to feature data 
# using tfd

vectorizer = TfidfVectorizer()
```

```python
feature_vector = vectorizer.fit_transform(combined_feature)
```

```python
print(feature_vector)
```

```python
# to find the similarity score
# using cosine similarity 

similarity = cosine_similarity(feature_vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
# now we will ask user to give the input
# and find which movie will be similar

movie_name = input('Enter your favourite Movie Name : ')
```

```python
# creating a list with all the movies given in the dataset

list_of_all_titles = movies_data['title'].tolist()
```

```python
list_of_all_titles
```

```python
# finding the close match for the movie name given by the user

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
```

```python
close_match
```

```python
# find the index of the movie with title

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]
```

```python
index_of_the_movie
```

```python
# take this index value find similar movies list

similarity_score = list(enumerate(similarity[index_of_the_movie]))
```

```python
similarity_score
```

```python
len(similarity_score)
```

```python
# finding highest similarity movies 

# sorting the movies based on their similarity score

sorted_similar_movies = sorted(similarity_score, key = lambda x : x[1], reverse = True)
```

```python
sorted_similar_movies
```

```python
# print the name of similar movies based on the index 

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i < 20):
        print(i, '.', title_from_index)
        i+=1
```

```python
# Movie REcommendation System
# copy pasting and arranging

movie_name = input('Enter the name of your favourite movie : ')

list_of_all_titles = movies_data['title'].tolist()

find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)

close_match = find_close_match[0]

index_of_the_movie = movies_data[movies_data.title == close_match]['index'].values[0]

similarity_score = list(enumerate(similarity[index_of_the_movie]))

sorted_similar_movies = sorted(similarity_score, key = lambda x : x[1], reverse = True)

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
    index = movie[0]
    title_from_index = movies_data[movies_data.index == index]['title'].values[0]
    if(i < 20):
        print(i, '.', title_from_index)
        i+=1
```
