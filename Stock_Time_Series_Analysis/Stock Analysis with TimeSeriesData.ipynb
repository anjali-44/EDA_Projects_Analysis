## Time Resamplling

```python
df_tesla.head()
```

```python
# A -> year and frequency of last day of year
df_tesla.resample(rule = 'A').min()

# entire information is gettnig retrieved based on minimum throughout the entire year on last day of evry year
```

```python
df_tesla.resample(rule = 'A').max()
```

```python
df_tesla.resample(rule = 'A').min()['Open'].plot()
```

```python
df_tesla.resample(rule = 'A').max()['Open'].plot()
```

```python
# to quaterly start frequency

df_tesla.resample(rule = 'QS').max()
```

```python
df_tesla.resample(rule = 'QS').max()['High'].plot()
```

```python
# business and frequency

df_tesla.resample(rule = 'BA').max()['High'].plot()
```

```python
df_tesla.resample(rule = 'BA').max()
```

```python
df_tesla.resample(rule = 'BQS').max()['High'].plot()
```

```python
# plotting

df_tesla['Open'].resample(rule = 'BA').mean().plot(kind = 'bar')
```

```python
df_tesla['Open'].resample(rule = 'A').mean().plot(kind = 'bar')
```

```python
df_tesla['Open'].resample(rule = 'M').max().plot(kind = 'bar')
# M -> monthly mean
```

```python
df_tesla['Open'].resample(rule = 'M').max().plot(kind = 'bar', figsize = (15,6))
```
